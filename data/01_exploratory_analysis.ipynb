```python
# wine_quality_analysis.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Carregar o dataset
dfwine = pd.read_csv("wine+quality/winequality-red.csv", sep=';')
```

```python
# 2. Visão geral
print("\nDimensões:", dfwine.shape)
print("\nColunas:", dfwine.columns.tolist())
print("\nAmostra dos dados:")
print(dfwine.head())
```

```python
# 3. Renomear as variáveis
dfwine.columns = ['FA','VA','CA','RS','CL','FSD','TSD','D','pH','S','OH','Q']
print(dfwine.head())
```

```python
# 4. Estatísticas descritivas
print("\nEstatísticas descritivas:")
print(dfwine.describe())
```

```python
# 5. Distribuição da variável alvo (qualidade)
plt.figure(figsize=(8, 5))
sns.countplot(x="Q", hue="Q", data=dfwine, palette="RdPu", legend = False)
plt.title("Distribuição da Qualidade do Vinho")
plt.xlabel("Qualidade (0–10)")
plt.ylabel("Frequência")
plt.tight_layout()
plt.savefig("quality_distribution.png")
plt.show()
```

```python
# 6. Matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(dfwine.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação entre Variáveis")
plt.tight_layout()
plt.savefig("correlation_matrix.png")
plt.show()
```

```python
# 7. Análise simples: qualidade vs álcool
plt.figure(figsize=(8, 5))
sns.boxplot(x="Q", y="OH", hue = "Q", data=dfwine, palette="muted", legend = False)
plt.title("Distribuição de Álcool por Nível de Qualidade")
plt.tight_layout()
plt.savefig("alcohol_vs_quality.png")
plt.show()
```
